# **A KNN**
**How many distances you need to calculate if you have 60,000 samples in
the trainingset for 50 samples?**

Es wird die Entfernung von unserem Sample (v) zu allen anderen Samples in unserem Trainingsset (M) berechnet.

AnzahlBerechnungen = v * M = 60000 * 50 = 3000000


**How many distances do you need to calculate if you have n samples im
the trainingset?**

Anstatt der vorherigen 60000 Samples in Trainingsset haben wir nun n Samples. Nun wird von unserem Sample (v) die Entfernung zu den n Samples in unserem Datenset berechnet. Mit n Samples ergibt sich somit eine Anzahl von: 

AnzahlBerechnungen = v * M = 50 * n = 50n
